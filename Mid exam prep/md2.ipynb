```python
import numpy as np
import math
import matplotlib.pyplot as plt 
import math
from sympy import symbols,diff
```

```python
x = symbols('x')
f_x = x**5+x**4
n =2 
f_n = diff(f_x,x,n)
f_n
```

```python
def taylor_series_cos(x,order):
    approx = np.zeros_like(x)
    for k in range((order//2)+1):
        approx = approx + ((-1)**k * x**(2*k)) / math.factorial(2*k)
    return approx
```

```python
x = np.linspace(-np.pi, np.pi, 200)
orders = [0,2,4,6]


for order in orders:
    y = taylor_series_cos(x,order)
    plt.plot(x,y)
```

```python
test_val =1.5
true_val = np.cos(1.5)

test_val_approx={}
test_val_errors={}

for order in orders:
    val = taylor_series_cos(test_val,order)
    test_val_approx[order] = val
    test_val_errors[order] = abs(true_val- val)

    print(f'Value for {order} Order Taylor series approximation at x = {test_val}: {test_val_approx[order]}')
    print(f'Error for {order} Order Taylor series approximation at x = {test_val}: {test_val_errors[order]}')


```

```python
def f(x):
    return x**3 -x-1

a,b = 1,2

def bisection(f,a,b,tol =0.1):
    m = (a+b)/2
    if np.abs(f(m)) <tol:
        return m
    elif np.sign(f(a)) == np.sign(f(m)):
        return bisection(f,m,b,tol)
    elif np.sign(f(b)) == np.sign(f(m)):
        return bisection(f,a,m,tol)

print(bisection(f,1,2,0.1))
```
