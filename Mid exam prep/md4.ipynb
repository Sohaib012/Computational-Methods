```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [0, 2, 4, 6, 8]
y = [0, 5, 4, 2, 3]
```

```python
n = 5
xp = np.linspace(-1,9,500)

yp = 0

for i in range(n):
    p=1
    for j in range(n):
        if i !=j:
            p = p * (xp-x[j])/(x[i]-x[j])
    yp = yp + p* y[i]
    
```

```python
plt.plot(xp,yp)
plt.scatter(x,y)
```
