```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
f = lambda x: np.cos(x) -x
g = lambda x: np.cos(x)

p_arr = []

TOL = 0.0001
N = 100
i =1 
p0 =0
p =0
while i<=N:
    p = g(p0)
    p_arr.append(p)
    if abs(p-p0) <TOL:
        break
    i = i+1
    p0 =p
    
```

```python
x = np.linspace(-1,2,400)
f_x= f(x)
g_x = g(x)

p_arr = np.array(p_arr)

plt.subplot(1,2,1)
plt.plot(x, f_x, label="f(x) =  $\\cos(x) - x$")
plt.scatter(p_arr, f(p_arr), label="P", color="red")
plt.subplot(1,2,2)
plt.plot(x, g_x, label="f(x) =  $\\cos(x) - x$")
plt.scatter(p_arr, g(p_arr), label="P", color="red")
```

```python
f = lambda x: np.exp(x) - 2
df = lambda x: np.exp(x)
```

```python
x0 = 0.5
TOL = 0.0001
N=100

x_vals = [x0]

for n in range(N):
    f_x0 = f(x0)
    df_x0 = df(x0)

    if df_x0 == 0:
        print('Np derivative')
        break
    x1 = x0-f_x0/df_x0
    x_vals.append(x1)

    if abs(x1-x0) <TOL:
        break
    
    x0 = x1
```

```python
x = np.linspace(-1,2,400)
f_x= f(x)
g_x = g(x)

x_arr = np.array(x_vals)

plt.subplot(1,2,1)
plt.plot(x, f_x, label="f(x) =  $\\cos(x) - x$")
plt.scatter(x_arr, f(x_arr), label="P", color="red")

```
