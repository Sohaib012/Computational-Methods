```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [0, 2, 4, 6, 8]
y = [0, 5, 4, 2, 3]
```

```python
# Reading number of unknowns
n = 5

# Reading interpolation point
xp = np.linspace(-1,9,500)

# Set interpolated value initially to zero
yp = 0

# Implementing Lagrange Interpolation
for i in range(n):
    
    p = 1
    
    for j in range(n):
        if i != j:
            p = p * (xp - x[j])/(x[i] - x[j])
    
    yp = yp + p * y[i]
```

```python
plt.figure(figsize=(10,6))
plt.title("Lagrange Interpolation Method")
plt.xlabel('$x$-axis')
plt.ylabel('$y$-axis')
plt.plot(xp, yp, label="Lagrange Interpolation")

plt.scatter(x, y, label="Original Values", color="red", marker="x")

plt.legend()
plt.grid()
plt.show()
```
