```python
def taylor_cos(x, order):
    y_approx = np.zeros_like(x)
    for k in range((order) // 2 +1):
        term = ((-1) ** k * x ** (2 * k)) / math.factorial(2 * k)
        y_approx += term
    return y_approx
```

```python
test_val = 1.5
true_val = np.cos(test_val)
orders = [0, 2, 4, 6]
```

```python

x = np.linspace(-np.pi, np.pi, 200)
test_val_approx = {}
test_val_errors = {}

plt.figure(figsize=(10, 6))

for order in orders:
    y_approx = taylor_cos(x, order)
    plt.plot(x, y_approx, label=f'Order {order}')

    val = taylor_cos(np.array([test_val]), order)
    test_val_approx[order] = val
    error = abs(true_val - val)
    test_val_errors[order] = error
    
    

# Plot the true cosine function
plt.plot(x, np.cos(x), 'k--', label='True cos(x)')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Taylor Series Approximation of cos(x)')
plt.legend()
plt.grid(True)
plt.show()

print("Approximations:", test_val_approx)

print("Errors:", test_val_errors)
```
