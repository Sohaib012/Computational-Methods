```python
# Task 1

import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad 
```

```python
def my_func(x):
  return np.cos(x)+2
```

```python
n = np.linspace(-np.pi,np.pi,100)
```

```python
f_X = my_func(n)
```

```python
f_X
```

The following graph is the plot from -pi to pi with green color and dashed line

```python
plt.plot(n,f_X,'g--')
plt.xlabel('x-values')
plt.ylabel('Amplitude')
plt.legend('f(X)')
plt.grid()
```

```python
def integrand(x):
  return np.cos(x)
```

```python
a = -np.pi
b = np.pi
n =10
delta = (b-a)/(n-1)
x = np.linspace(a,b,n)
```

```python
x
```

```python
actual_val,_ = quad(integrand,a,b)
```

```python
print('actual_value',actual_val)
```

```python
print((x[:n-1] + x[1:]) / 2)
```

```python
x_mid =(x[1:] + x[:-1]) / 2
```

```python
reimann_mid = delta * np.sum(np.cos( (x[:n-1] + x[1:]) / 2))
```

```python
reimann_mid
```

```python
y = delta *np.cos( (x[:n-1]) )+ np.cos(x[1:]) / 2
```

```python
y
```

```python
err = actual_val-reimann_mid
```

for n=10 the error is as follows, firstly i computed the quad the acutal value for the integration using scipy libraary. 
Then computed the mid points of y using y = delta *np.cos( (x[:n-1]) )+ np.cos(x[1:]) / 2 in which y values are computed and their mid is taken. Then finally i plotted using matplotlib.pyplot and using a bar chart showcased the midpoints 

```python
print("Error for n=10",err)
```

```python
plt.plot(x_mid,y)
plt.bar(x_mid,y,width=0.7,fill=False)
plt.xlabel('x-values')
plt.ylabel('Amplitude')
plt.legend('f(X)')
plt.grid()
```

for n=100 the error is as follows, firstly i computed the quad the acutal value for the integration using scipy libraary. 
Then computed the mid points of y using y = delta *np.cos( (x[:n-1]) )+ np.cos(x[1:]) / 2 in which y values are computed and their mid is taken. Then finally i plotted using matplotlib.pyplot and using a bar chart showcased the midpoints 

```python
a = -np.pi
b = np.pi
n =100
delta = (b-a)/(n-1)
x = np.linspace(a,b,n)

actual_val,_ = quad(integrand,a,b)
x_mid =(x[1:] + x[:-1]) / 2
reimann_mid = delta * np.sum(np.cos( (x[:n-1] + x[1:]) / 2))
y = np.cos( (x[:n-1]) )+ np.cos(x[1:]) / 2

err = actual_val-reimann_mid
print("Error for n=100",err)
```

```python
plt.plot(x_mid,y)
plt.bar(x_mid,y,width=0.07,fill=False)
plt.xlabel('x-values')
plt.ylabel('Amplitude')

plt.grid()
```
